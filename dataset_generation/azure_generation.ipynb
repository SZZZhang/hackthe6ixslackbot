### Generating the rating matrix

```python
# generating the pairs
pair_sets = set([])
pairs = []
for i in range(5000):
    sample = tuple(random.sample(ids, 2))
    sample2 = (sample[1], sample[0])
    if sample not in pair_sets:
        pairs.append(sample)
        pairs.append(sample2)
        pair_sets.add(sample)
        pair_sets.add(sample2)
```

```python
len(pairs)
```

```python
number_encounters = len(pairs)
userid1 = [i[0] for i in pairs]
userid2 = [i[1] for i in pairs]
rating = []
weights = []


for i in range(int(number_encounters/2)):
    response1 = new_df.loc[userid1[i], ["Q" + str(i) for i in range(1,11)]]
    response2 = new_df.loc[userid2[i], ["Q" + str(i) for i in range(1,11)]]
    differences = abs(response1 - response2).sum()
    if differences > 7:
        sample = 0
    elif differences > 4:
        sample = 1
    else:
        sample = 2
    
    rating.append(sample)
    rating.append(sample)
```

```python
rating_df = pd.DataFrame()
rating_df["userid1"] = userid1
rating_df["userid2"] = userid2
rating_df["rating"] = rating
```
